```python
import matplotlib.pyplot as plt
anyo=11
genero='M'
top=5
df['id']=df.Nombre+"-"+df.Club
df2=df[(df.Anyo_nac==anyo) & (df.M_F==genero)][['Nombre','Club','Prueba','Puesto','id']]
# df2=df[(df.Anyo_nac==anyo) & (df.M_F==genero)].loc[:,['Nombre','Club','Prueba','Puesto']]

df4=pd.concat([df2[df2.Prueba==j].sort_values(by='Puesto',ascending=True).head(top) for j in df2.Prueba.unique()])

na = df4.id.unique()

fg, ax = plt.subplots()
ax.set_ylabel('Puesto')
ax.set_title('Evolucin de puestos')
plt.yticks(range(1,top+1))


for n in na:
    df5=df4[df4.id==n]
    plt.plot(df5.Prueba, df5.Puesto, label=n)
    plt.scatter(df5.Prueba, df5.Puesto)
    
plt.grid(True, axis='y')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0)
plt.show()
```

```python
df2=df[['Club','Prueba','Puesto']]
fg, ax = plt.subplots()
plt.boxplot([df2[df2.Club==c].Puesto for c in df2.Club.unique()],labels=df2.Club.unique())
plt.xticks(rotation=90)
ax.set_ylabel('Puesto')
ax.set_title('Distribucin de puestos por club')
plt.show()
```

```python
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(9, 4))

# Fixing random state for reproducibility
np.random.seed(19680801)


# generate some random test data
all_data = [np.random.normal(0, std, 100) for std in range(6, 10)]

# plot violin plot
axs[0].violinplot(all_data,
                  showmeans=False,
                  showmedians=True)
axs[0].set_title('Violin plot')

# plot box plot
axs[1].boxplot(all_data, labels=['x1', 'x2', 'x3', 'x4'])
axs[1].set_title('Box plot')

# adding horizontal grid lines
for ax in axs:
    ax.yaxis.grid(True)
#     ax.set_xticks([y + 1 for y in range(len(all_data))], labels=['x1', 'x2', 'x3', 'x4'])
    ax.set_xlabel('Four separate samples')
    ax.set_ylabel('Observed values')

plt.show()

all_data
```
